```python

# Step 2: Import libraries
import os
from getpass import getpass
from groq import Groq

# Step 3: Set API Key securely
os.environ["GROQ_API_KEY"] = getpass("Enter your Groq API Key: ")

# Step 4: Initialize client
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))
```

```python

#  LLM Multi-Agent Support System (Colab Ready)
#  LLM CALL FUNCTION
def call_llm(prompt):
    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": "You are a helpful customer support assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"Error in LLM call: {e}"

# AGENTS

# 1. Intent Classifier Agent
def classify_intent(query):
    prompt = f"""
    Classify the following user query into ONE category:
    - billing
    - technical
    - general

    Only return the category name.

    Query: {query}
    """
    return call_llm(prompt).lower()


# 2. Billing Agent
def billing_agent(query):
    return call_llm(f"User has a billing issue. Help resolve: {query}")


# 3. Technical Agent
def technical_agent(query):
    return call_llm(f"User has a technical issue. Help resolve: {query}")


# 4. General Support Agent
def general_agent(query):
    return call_llm(f"User has a general question. Help: {query}")

# CONTROLLER (ROUTER)
def run_system(user_query):
    intent = classify_intent(user_query)

    if "billing" in intent:
        return billing_agent(user_query)
    elif "technical" in intent:
        return technical_agent(user_query)
    else:
        return general_agent(user_query)


# MAIN LOOP
print("🚀 LLM Multi-Agent Support System Started (type 'exit')")

while True:
    user_input = input("\nEnter your query: ")

    if user_input.lower() == "exit":
        print("👋 Exiting system...")
        break

    output = run_system(user_input)
    print("\n🤖 Response:", output)
```
